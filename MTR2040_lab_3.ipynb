The main purpose of this lab is to introduce students to the basic fundamentals of plotting  in Python using the **matplotlib** library. We'll also get into the **pandas** library which is the most common package for storing, viewing, and manipulating tabular data.


```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
# create some dummy data
temperatures = [11, 11.6, 12.6, 15.7, 20, 22, 25, 26, 21, 16, 14, 10]
```

Using the **plt.plot()** funciton can immeadiatly plot a line graph of 1D data. If only one argument is provided, it will plot the data on the y-axis and use the index values for the x-axis.  

```python
# plot the data as a line graph
plt.plot(temperatures)
```

If you specify an x-axis, you can feed in both in the order (x, y). This can be a list of strings or numerics.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
plt.plot(months, temperatures)
```

Let check the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) to see what else we can do with it. There are a lot of options in the "kwargs" (key word arguments) that relate to style. Play around with a bunch of options.

```python
plt.plot(months, temperatures, color='red', linewidth=5, linestyle='--')
```

To plot more than one line, you can simply add another plot command and it will add it to the figure.

```python
dew_points = [8, 11, 6.67, 13.7, 10, 12, 16, 24, 20, 9, 3, 5]
plt.plot(months, temperatures, color='red', linewidth=5, linestyle='--')
plt.plot(months, dew_points, color='blue', linewidth=5, linestyle='--')
```

To add or change other labels, such as titles, labeling the axes or adding a legend, other methods are used (instead of key word arguments). Again, Google is your friend, here.

```python
plt.plot(months, temperatures, color='red', linewidth=5, linestyle='--', label='Temperature')
plt.plot(months, dew_points, color='blue', linewidth=5, linestyle='--', label='Dew Point')
plt.title('Seasonal Temperature and Dew Ponits', fontsize=16)
plt.ylabel('Degrees (f)', fontsize=12)
plt.xlabel('Month', fontsize=12)
plt.legend()
```

So far, we've populated some small lists manually for data, but this would quickly become unreasonable with even a little more data. Numpy can help us here. Numpy has many built-in functions to sample distributions. Let's start with sampling the [numpy normal distribution](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html).

```python
n_samples = 100
normal_dist = np.random.normal(size=n_samples)
print(normal_dist)
```

Let's plot a histogram to see if it looks "normal". Instead of `plt.plot()` we'll use the `plt.hist()` function.

```python
plt.hist(normal_dist)
```

Looks relatively normal ... but let's make it more interesting with more data and more bins.

```python
n_samples = 1000000 
normal_dist = np.random.normal(size=n_samples)
plt.hist(normal_dist, bins=100)
plt.show()
```

```python
n_samples = 1000000 
normal_dist = np.random.normal(size=n_samples)
plt.hist(normal_dist, bins=100)
plt.show()
```

If we look in the documentation, we can see that the `random.normal()` documentation has two other optional arguments - scale and location. Can we guess what the distribution will look like if we change these parameters from the default?

```python
n_samples = 1000000 
normal_dist = np.random.normal(size=n_samples)
second_dist = np.random.normal(size=n_samples, loc=5, scale=2)
plt.hist(normal_dist, bins=100, color='blue', alpha=0.5)
plt.hist(second_dist, bins=100, color='orange', alpha=0.5)
plt.show()
```

"object" in pandas is essentially means everything in that columns will be interpreted as a string - thus mathematical operations wouldn't work.

Let's see what happens when we try to plot the date column on the X-axis.

Well, technically, it did plot it. But it's not really what we're looking for. It plotted every single value as a string, but because there are so many values they are all overlapping each other. 

In Python, 'datetime64' (64bit time) is it's own data type, and is very useful for working with time series. By default, the pandas `read_csv()` function does not parse dates, so you have to explicitly tell it to do so. If you don't explicity parse it in the `read_csv()` function (as an argument - see below) it will automatically assign it to the "object" datatype. In fact, anytime pandas doesn't understand the data type, it will default to "object".

You can either change the datatype after you load the data, or tell it to do so as you load it. See the [documentation](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) for how to use the `parse_dates=` argument in the `read_csv()` function. As you can see in the documentation, some functions can have a LOT of arguments!

```python
df.dtypes
```

```python
# You can change the whole columns datatype using the "astype" method
df['Date and Time'].astype('datetime64')
```

```python
df.dtypes
```

Hmmm... looks like it was converted to a datetime64 object in the cell using the `.astype('datetime64')` method, but when we check our whole dataframe it still shows up as an object type. This is because we didn't explicitly override / set the new value,  which should be done using the `=` sign like a traditional variable.

```python
df['Date and Time'] = df['Date and Time'].astype('datetime64')
df.dtypes # has now been properly converted
```

```python
plt.plot(df["Date and Time"], df["Air Temp"])
```

Now when we plot it, it recognizes the datetime dtype and can plot it in a smarter way (by only plotting every few intervals, for example).

To parse the dates automatically (if the dates are a column of it own), just pass the column name to the `parse_dates` argument of the `read_csv()` function. It needs to be passed as a list, per the documentation.

```python
df = pd.read_csv("/Users/charlesbecker/Downloads/Fort_collins_Ag_2023-01-01_2023-01-05.csv", parse_dates=['Date and Time'])
df.dtypes # success
```

```python
plt.figure(figsize=(16, 10)) # use this to explicitly initialize your figure, you can then change it's size, etc
plt.plot(df['Date and Time'], df['Air Temp'], linewidth=3, color='r', label='Temp')
plt.plot(df['Date and Time'], df['Dewpoint'], linewidth=3, color='g', label='Dewpoint')
plt.title("Fort Collins Temperature and Dewpoint: 2023-01-01 through 2023-01-05", fontsize=22)
plt.xlabel('Date', fontsize=20)
plt.legend()
```
